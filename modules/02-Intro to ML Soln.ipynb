### 1. Definitions

#### Machine Learning (ML)
What is ML? The following two general definitions describe how we interpret ML in this book:
* Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed (Samuel A. 1959)
* Machine learning is a method of data analysis that automates analytical model building. It is a branch of artificial intelligence based on the idea that systems can learn from data, identify patterns and make decisions with minimal human intervention (SAS, 2022)

Note that these definitions emphasis on the ability to learn patterns in data with minimal human intervention, as well as on decision making. However, this still requires scientists to gather appropriate data sets (e.g. sample size, variables) and pick an appropriate model strategy and type. In order for ML to have an impact on decision making, results need to be communicated properly and model uncertainty should be known. Therefor, this book will teach various ways to assess model performance and visualize model outputs.

#### Statistical Learning (SL)
Throughout the book authored by James et al. (2021), the term SL is consistently utilized instead of ML. It is important to point out that the terms SL and ML are often used interchangeably. Others may argue that they are two different fields.

In this book we acknowledge that SL and ML highly overlap, as both deal with the analysis of data. Both fields deal with making predictions as well as the formalization of relationships between variables (inference). However, in SL often tries to capture the relationship between predictor variables and response in mathematical equations. In addition, the focus is more on and hypothesis testing, as assumptions such as normality, multicollinearity, and homoscedasticity are typical in statistical learning. ML algorithms, such as random forest, often have less strict assumptions and are optimized for larger datasets (for example, millions of observations and/or variables).

General definitions of SL include:
* Statistical learning refers to a vast set of tools for understanding data (James et al. 2021)
* Statistical learning theory is a framework for machine learning drawing from the fields of statistics and functional analysis. Statistical learning theory deals with the statistical inference problem of finding a predictive function based on data (Vapnik 1995)

#### Big Data
Big data is associated with machine learning. Massive data sets are currently available due to continuous data gathering and storing, possible due to advanced sensor and communication techniques. Think of social media tracking our visits of outdoor and recreation sites, as well as satellite constellations such as Landsat continuously collecting imagery of the earth.
Such data sets are too large or complex to be dealt with by traditional data-processing application and analysis with simple SL models. Big data needs to be processed on multiple CPU cores on a single or multiple machines. The data may also be unstructured or semi-structured, further complicating analysis (Jacops, 2009). ML approaches are better tailored towards such data sets.

#### 2.1 Descriptive Analytics
Descriptive analytics uses data summarization to uncover trends and patterns. Summaries are typically visual data representations such as pie charts, box plots, as well as interactive dashboards. Such data summaries are often required in formats such as journal articles or reports. Descriptive analytics answers questions like "What happened?" and "What are the key characteristics of the data?" For example, analyzing historical climate data to understand temperature and rainfall patterns in a specific region is a form of descriptive analytics. Note that descriptive analytics is often required before you can make predictions or prescriptions. In the next few chapters you will learn how to clean, reshape, summarize, and effectively visualize data.

#### 2.2 Predictive Analytics
Predictive analytics tries to understand what might happen in the future. It is possible to predict classes (e.g. land cover class), associated probability scores, as well as continuous values (e.g. temperature of biomass). Predictive analytics answers questions like "What is likely to happen?" and "What will be the future outcome?" Typically predictive models provide insights on data inferences, which means we can figure out which predictor variables had largest influence on model outcomes. For example, analyzing historical fish migration patterns and environmental factors to forecast fish populations is a form of predictive analytics.

Predictive analytics is not only used to make future predictions. Instead, it can be used to analyize large unstructured data sets that were collected in the past, such as satellite based image archives and data collected by self-driving cars. For example, we cannot tell proportions of dominant tree species in Garibaldi park from Landsat imagery directly. However, we can potentially map dominant tree species and predict their proportions through modelling involving field samples.

#### 2.3 Prescriptive Analytics
In prescriptive analytics, users link outputs from descriptive and predictive analytics to possible actions. Predictions, as well as ecological knowledge derived from data inferences, are used to advice on management decisions. However, this requires careful assessment and communication of model performance and uncertainty. Prescriptive analytics answers questions like "What should we do?" and "What is the best course of action?" For example, recommending optimal harvesting strategies for forestry management to balance timber production with long-term forest sustainability is a form of prescriptive analytics. By integrating outputs from descriptive analytics (such as understanding tree growth patterns and forest composition) and predictive analytics (such as predicting future timber yields and forest dynamics), prescriptive analytics can suggest specific harvesting strategies that optimize timber production. This allows decision-makers in forestry management to make informed choices about when, where, and how to conduct harvesting activities, taking into account factors such as ecological conservation, timber supply, and economic considerations. Best practices and ethics related to prescriptive use of ML model outputs will be discussed towards the end of this book.

#### 3.3 Flexibility and Interpretability of Models
In machine learning there is a trade-off between **flexibility and interpretability** of models and their results.
* Interpretability is the degree to which a human can understand the cause of a decision. The higher the interpretability of a model, the easier for a human to understand why a decision was made. In science, the "why" question is often more important than achieving the most accurate predictions possible (the what).
* Flexibility is the degree to which models are able to follow more complex trends in data sets. A simple linear regression model is inflexible, because it is not able to follow such trends. In addition, a flexible model will work for data sets with higher variance, large data sets, as well as data sets with unevenly spread data.

Let’s say we want to train a model to classify forest stands as being productive/non-productive (binary classification), and we going to use a forest inventory data set that contains information related to forest structure, disturbance history, historic weather conditions, and site/soil conditions. If we are purely concerned with accurately predicting productive/non-productive stands, we may choose to employ a model with high flexibility and low interpretability. Conversely, it is more likely that while we want our model to make accurate predictions, we will be most concerned with why some sites are productive and some are not. In this case, it is likely best to choose to employ a model with high interpretability, for example, to learn which variables influenced the prediction output the most.

In essence, supervised learning refers to approaches estimating <em>f</em>, where <em>f</em> is the function that connects input variables to an output variable. Some supervised learning algorithms are flexible whereas others are more restrictive. Restrictive algorithms can just estimate a few shapes of function <em>f</em>. #### 3.4 Parametric and Non-parametric Models
Machine learning models can be characterized as **parametric and non-parametric**. Parametric models make an assumption about the functional form, or shape, of f. For example, when we use a linear model we assume f is linear. This is followed by the training procedure, during which the model parameters are determined. A common way to do this is through ordinary least squares.

Non-parametric models do not make explicit assumptions about the form of function <em>f</em>. This typically allows them to better fit the functional form of <em>f</em> and makes them more flexible with regard to which data sets they can be applied to. However, this also makes non-parametric models harder to interpret. In addition, non-parametric models require a larger number of observations <em>f</em>, which arguably limits their flexibility.

Figure 5 illustrates that inflexible (e.g. linear and logistic regression) as well as more flexible models (e.g. generalized additive models) can be parametric. Non-parametric models include models based on decision trees (e.g. random forests) and neural networks (i.e. deep learning). Deep learning is highlighted in the figure as it is often, wrongly, described as a parametric model due to it's typically high number of model parameters.

### 4. Generalization, Overfitting, and Underfitting

Ultimately, we want to train models on a representative data set so that they perform well when predicting on data they were not trained with. For example, we might train a model on plot data from several stands which collectively represent the full range of conditions in a specific ecoregion. Our goal for this model would then be to apply it to make predictions across this entire ecoregion so that we can characterize areas where we have not collected plot data. If the model can achieve this, we can say that the model generalized well. Conversely, if the model only performs well on areas where plot data were collected, it has likely overfit the training data. If the model performs poorly even on the training data, it has underfit the training data.

* **Generalization:** The ability of a model to generalize from a training dataset and make accurate predictions on "unseen" data.
* **Underfitting:** When a model is too simple to capture all the aspects of and variability in the dataset, and as a result, makes inaccurate predictions, even on the training dataset. We can try the following to mitigate underfitting and develop machine learning models that capture the complexity of the data, leading to improved predictive performance and better generalization:<br>
    * Incorporate additional relevant features to provide the model with more informative inputs. This can involve transforming variables, creating interaction terms, or including subject-specific knowledge.
    * Increase training data to learn more complex relationships and improve generalization
    * Combine multiple models through ensemble methods, such as bagging, boosting, or stacking
    * Perform Cross-Validation to evaluate model performance on multiple subsets of the data

* **Overfitting:** Overfitting happens when the model is too complex. In this case it fits the noisiness of the training dataset too well, and as a result makes inaccurate predictions on "unseen" data. There are several ways to avoid overfitting: <br>

    * Gather more training data
    * Reduce noise in the training data
    * Select a model with fewer parameters (e.g. linear model instead of a high-degree polynomial model)
    * Reducing the number of predictor variables
    * Constraining the model, which is called regularization.

### 5. Model Performance for Classification Tasks

A definition of machine learning:
A computer program is said to learn from experience E with respect to some task T
and some performance measure P, if its performance on T, as measured by P, improves
with experience E.
—Tom Mitchell, 1997

So what is a performance measure? There are many performance measures, however, selecting the appropriate measures and correctly interpreting them is challenging. Proper selection of measures is necessary.

#### 5.1 Confusion Matrix
Most performance measures are based on the number of correct and incorrect classified instances. We can check how many times the model correctly and incorrectly predicted the presence of a particular class (true and false positives), as well as where the model correctly and incorrectly predicted the absence of a particular class (true and false negatives).

Here is an overview of these terms for a binary present-absence model.
* **True Postive (TP):** The model correctly predicts the positive class
* **False Positive (FP):** The model incorrectly predicts the positive class
* **True Negative (TN):** The model correctly predicts the absence class
* **False Negative (FN):** The model incorrectly predicts the absence class

Correct and incorrect classified instances are typically visualized using a confusion matrix. Figure 7 shows the most common format of a confusion matrix.

Figure 8 shows a confusion matrix for a model, based on a remotely sensed data set, predicting which pixels are dominated by conifer or deciduous tree species.

The matrix shows that the number of correctly predicted conifer dominant pixels (True Positives and True Negatives) is considerably larger than the incorrectly predicted conifer dominant pixels (False Positives and False Negatives). Still, the quality of the model is still hard to assess due to class imbalance. We have 2521+119 conifer pixels, while we have 68016+751 deciduous pixels. If we turn the absolute numbers into percentages it will become much easier to interpret! Therefore, the true positive rate, true negative rate, false postive rate, and false negative rate were calculated and added to the confusion matrix. The equations are listed here but you do not have to worry about them at this stage...

* True positive rate (TPR) is the ratio between the number of True Postives against the total number of Positives. This can be calculated by <br>  *TP / (TP + FN)*.
* True negative rate (TNR) is the ratio between the number of True Negatives against the total number of Negatives. This can be calculated by <br> *TN / (TN + FP)*.
* False positive rate (FPR) is the ratio between the number of False Positives against the total number of Negatives. This can be calculated by <br> *FP / (TN + FP)*.
* False negative rate (FNR) is the ratio between the number of False Negatives against the total number of Postives. This can be calculated by <br> *FN / (TP + FN)*.

#### 5.3 Recall, Precision and F-score
We have seen that accuracy is not always a good metric to use, particularly when having classes of different sample sizes (i.e. class imbalance). So what metrics can we use instead? Let's look at the metrics Recall and Precision.

Figure 10 shows how recall and precision perform with different class sample sizes. Recall is mostly influenced by the number of false negatives of the concerning class. For example, in both panels the Recall of class A is relatively low (0.7), because we have 3 false negatives while having 10 positives total. Precision is influenced by how many false positives are present within the concerning class. For example, in the first panel the Precision of Class A is 0.875 because there is 1 false positive found while have 7 true positives.

Finally, the F-score is the harmonic mean of the precision and recall.

#### 5.4 Cross-validation
In supervised machine learning we use only part of the sample data for training (e.g. 80%) and the remainder for testing. Performance measures are calculated based on the test data sets. Why do we reserve some for testing? Because the samples used to calculate performance measures cannot have been "seen" by the model, this will provide too positive performance measures which do not represent the model's actual performance in "new" situations. Can it be challenging to select which data will be reserved? Yes, as the samples for the test data set are randomly drawn the derived performance measures can highly vary. How does cross validation make this easier? During cross validation, train and test samples are drawn <em>n</em> times from the full sample data set, thus creating the same <em>n</em> number of models. This allows us to also calculate the performance measures <em>n</em> times and compare these.

A popular cross-validation method is **K-fold cross-validation**. Figure 11 illustrates the data set is split into k-folds, in this case k=5. During cross-validation, each fold is used once as a test data set, while the other four folds are used as training data. This is repeated <em>k</em> times.

### 6. Model Performance for Regression Tasks

For most applications using the standard error or R-squared as the scoring method is enough. However, R-squared can only be used for linear regression models. The standard error of the regression can be used for non linear models.
In addition to these metrics, cross-validation is a valuable technique that can be employed to assess the performance of regression models. Cross-validation involves partitioning the data into multiple subsets, training the model on a subset, and evaluating it on the remaining data. This repeated process provides a more robust estimate of the model's generalization capabilities and helps identify potential issues such as overfitting or underfitting.

 Therefore, the performance metrics commonly used in regression tasks include:

* Residual Standard Error (RSE): indicates how far the data points are from the regression line on average. The lower the RSE the better.
* R-squared: indicates what percentage of the variance is explained by the regression line.
* Cross-validation

```python
### Load relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

### Read and edit data
df = pd.read_csv("/content/drive/MyDrive/FRST505/week2/covtype-2.csv") # Read

rename = {1:'conifer', 2:'conifer', 3:'conifer', 6:'conifer', 4:'other', 5:'other', 7:'other'} # Rename class labels
df = df.replace({'Cover_Type':rename})

df.dropna(inplace=True)

X = df.loc[:, ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', # X: Select predictor variables
               'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways', 'Hillshade_9am',
               'Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points']]

y = df.loc[:, 'Cover_Type'] # y: Select column with class labels
```

```python
### Fit and apply a Random Forest Classifier
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8) # Split in training and validation dataset

rf = RandomForestClassifier(n_jobs=-1) # Initiate random forest classifier
rf.fit(X_train, y_train) # Fit
y_pred = rf.predict(X_val) # Make predictions on the validation dataset
```

```python
### Create confusion matrices
matrix_normalize = confusion_matrix(y_val, y_pred, normalize="true")
```

```python
### Plot confusion matrix
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(7,7)) # Initialize plot
fig.suptitle(" ")

labels = ['TN: ' + str(matrix_normalize[0][0]) + '\n \n TNR: ' + str(round(matrix_normalize[0][0]*100, 2)) +'%', # Create labels
         'FP: ' + str(matrix_normalize[0][1]) + '\n \n FPR: ' + str(round(matrix_normalize[0][1]*100, 2)) +'%',
         'FN: ' + str(matrix_normalize[1][0]) + '\n \n FNR: ' + str(round(matrix_normalize[1][0]*100, 2)) +'%',
         'TP: ' + str(matrix_normalize[1][1]) + '\n \n TPR: ' + str(round(matrix_normalize[1][1]*100, 2)) +'%']
labels = np.asarray(labels).reshape(2,2) # Reshape labels in 2x2 array

sns.heatmap(matrix_normalize, annot=labels, fmt='', cmap='Blues', ax=axs) # Create confusion matrix plot here

plt.tight_layout() # Finalize plot
plt.show()
```

Overall, this code aims to train a random forest classifier on the "covtype.csv" dataset and evaluate its performance using a confusion matrix. The confusion matrix provides insights into true negatives (TN), false positives (FP), false negatives (FN), and true positives (TP). The visualization helps assess the model's ability to correctly classify the different cover types based on the selected predictor variables.

### 10. Extra Exercise: Overfitting Exploration

**Objective:** Experiment with model complexity to understand overfitting and generalization.

**Instructions:**

1. Train three Random Forest models with different numbers of trees (n_estimators = 10, 100, 500).

2. Compare their performance on both training and validation data.

3. Plot the accuracy or F1-score for each model and observe trends.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

n_trees = [10, 25, 50, 75, 100]
train_scores, val_scores = [], []

# Sample smaller portion (70%) for speed ~2 minutes 40 seconds
X_sub = X_train.sample(frac=0.7, random_state=42)
y_sub = y_train.loc[X_sub.index]

for n in n_trees:
    rf_temp = RandomForestClassifier(
        n_estimators=n,
        max_depth=10,
        n_jobs=-1,
        random_state=42
    )
    rf_temp.fit(X_sub, y_sub)
    train_pred = rf_temp.predict(X_sub)
    val_pred = rf_temp.predict(X_val)

    train_scores.append(f1_score(y_sub, train_pred, pos_label="conifer"))
    val_scores.append(f1_score(y_val, val_pred, pos_label="conifer"))

plt.plot(n_trees, train_scores, label="Training F1")
plt.plot(n_trees, val_scores, label="Validation F1")
plt.xlabel("Number of Trees")
plt.ylabel("F1-Score")
plt.legend()
plt.title("Effect of Model Complexity on Overfitting ")
plt.show()
```

4. Discuss:

- Does performance continue improving with more trees?

- Is there evidence of overfitting or underfitting?

Yes, performance generally continues to improve with more trees, but the rate of improvement sharply decreases after about 50 trees.
In practice, increasing the number of trees beyond 50 offers diminishing returns in F1-score while significantly increasing computation time.

There is clear evidence of mild overfitting as the number of trees increases.

- Evidence of Overfitting: Overfitting occurs when a model performs significantly better on the training data than on the validation data.
    - The Training F1 curve is consistently higher than the Validation F1 curve across the entire range of the graph (10 to 100 trees).
    - The gap between the two curves starts small at 10 trees (difference $\approx 0.0005$) and widens to a peak difference of approximately $\approx 0.0007$ or $\approx 0.0008$ around 50-100 trees. This widening gap demonstrates that the model is learning the training data's noise and specific patterns too well, causing a slight degradation in its ability to generalize to new data.
- Evidence of Underfitting: There is no evidence of underfitting in the displayed range. Underfitting would be characterized by both the training and validation scores being very low, which is not the case here, as both F1-scores are very high ($\ge 0.981$). The model is performing very well overall.
