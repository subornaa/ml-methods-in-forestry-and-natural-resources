#### Estimating merchantable stem volume

<img src=https://www.for.gov.bc.ca/ftp/DCO/external/!publish/RecSitesTrails/SIGN_MAKER_RESOURCES/RSTBC_%20New_Logos/MFLNRO_%20no%20RSTBC.jpg width="400">
Since 1976 several equations have been developed and maintained by the British Columbia Ministry of Forests, Lands and Natural Resource Operations for modelling individual tree volume. [Nigh (2016)](https://www.for.gov.bc.ca/hfd/pubs/docs/TR/TR106.pdf) developed new models enabling the prediction of total tree volume and merchantable volume for British Columbia. Note that merchantable tree volume is total tree volume minus the volume of the stump and top. The four models focus on:

* total volume by species and region (coast or interior)
* total volume by species and biogeoclimatic zone; 
* merchantable volume by species and region (coast or interior)
* merchantable volume

For the development of the new models, trees of major commercial species were felled and sectioned. Based on stem diameters measured at each cross section total and merchantable volume are derived. Trees were sectioned at 0.3 (stump height), 0.6, 0.9, 1.37 (breast height) meters, as well as ten equally spaced points above 1.37 meter. In addition, tree height, species, region and biogeoclimatic zone were recorded. For each biogeoclimatic zone-species combination, at least 30 tree samples were used for modelling, however sample sizes were often larger than 500.
 
* Discussion: Which Machine Learning techniques could help build models predicting volume? How would you incorporate the described datasets?

* G.D. Nigh (2016) "Total and merchantable volume equations for common tree species in British Columbia by region and biogeoclimatic zone." Technical Report-Ministry of Forests, Lands and Natural Resource Operations, British Columbia. https://www.for.gov.bc.ca/hfd/pubs/Docs/Tr/Tr106.htmpip

#### Characterize the visitors of protected areas
<img src=https://www.dreamoverland.com/wp-content/uploads/2020/09/mapa_APs_site_2-760x1024.jpg width="450">
Nature-based tourism, which depends on natural outdoor environments as the main setting for tourist activities, is a fast growing economic activity globally. It has large social-economic potential in countries like Portugal, where internal tourism consumption is over 10 percent of the GDP and international tourism reached 7.4 billion Euros revenue. At the same time, in Portuguese society there has been a growing concern about environmental issues related to leisure activities in general. The Portuguese government has set rules and restrictions for tourism in protected areas but also promoted nature-based tourism contributing to development of local rural communities.

Currently there is a lack of information on the characterization of visitors of Portugal's protected areas and their motivations. Visitor characterization could help identify target markets and help develop travel destinations (including infrastructure, communication strategies, rules, restrictions) while including conservation priorities. Benefit segmentation is an approach that can help partition consumer markets in more homogeneous, and better interpretable, clusters. Within clusters, consumers are likely exhibit similar behavior. Clusters can be formed by socio-demographic, geography and behavioral (e.g. benefits, preferences) attributes. These attributes are commonly gathered using a self-administered questionnaire.

We just summarized part of the introduction of [Marques et al. (2010)](https://doi.org/10.1080/09669582.2010.497222). They have distributed a questionnaire under people living in the Portuguese mainland aged 15 to 75 years able to visit protected areas, which was filled out by 779 individuals. The questionnaire was partitioned into five sections measuring aspects such as socio-demographic characteristics and psychographics, occupation of leisure time and travel habits, environmental attitudes.

* Discussion: Which Machine Learning techniques could help with visitor characterization using the described survey dataset and "benefit segmentation" approach?

#### Predicting land cover type
<img src=https://irss.sites.olt.ubc.ca/files/2016/08/annual_proxy.gif width="500">

Land cover refers to the observed biotic and abiotic composition of Earth’s surface, which includes vegetation and anthropogenic structures. Mapping land cover is critical to understand and monitor the status and conditions of ecosystems. On top of that, land cover maps can be an input for modelling of carbon storage, biodiversisity, habitat, and resource management activities.

Remote sensing has emerged as a mature technology for mapping land cover across a range of spatial scales. With Landsat satellite imagery this can be done at regional and global scales, at high spatial (30 meters) and temporal resolutions (16 days revisit time). The underlying assumption is that Landsat’s surface reflectance values are similar for features presenting the same physical characteristics, which allows land cover classification. Satellite images are covered by clouds. [Hermosilla et al. (2015)](doi:10.1016/j.rse.2014.11.005) has developed a method to create annual cloud-free Landsat images for all of Canada. 

During the first two decennia since 2000, the Earth Observation for Sustainable Development of forests (EOSD) land cover map has been leading in Canada. EOSD land cover is based on Landsat imagery as well, however pixels were grouped using an unsupervised clustering approach followed by extensive manual editing by photo-interpreters. In other words, updating the EOSD land cover map would most likely be an expensive multi-year project. The EOSD map covers eight vegetation classes and four non-vegetation classes. Vegetation classes include four treed (wetland-treed, coniferous, broadleaf and mixedwood) and four non-treed (bryoids, herbs, wetland, and shrubs) classes.

This case study has been based on [Hermonsilla et al. (2018)](https://doi.org/10.1080/07038992.2018.1437719)

* Discussion: Which Machine Learning techniques could help map the described land cover classes accross Canada anually? What datasets could be used?

* T. Hermosilla,  M.A. Wulder, J.C. White, N.C. Coops & G.W. Hobart "Disturbance-informed annual land cover classification maps of Canada's forested ecosystems for a 29-year landsat time series." Canadian Journal of Remote Sensing, 44(1), pp. 67-87. https://doi.org/10.1080/07038992.2018.1437719
